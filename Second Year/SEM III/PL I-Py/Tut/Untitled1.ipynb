```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r'/content/laptop_price_dataset.csv')
weight = data['Weight']
price = data['Price']
fig=plt.figure(figsize=(200,70))
plt.scatter(weight, price, c="blue")
plt.title("Weight v/s Price comparison of laptops")
plt.xlabel("Weight")
plt.ylabel("Price")
plt.show()
```
