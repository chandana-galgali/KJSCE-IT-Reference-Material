**TEXT ENCRYPTION AND DECRYPTION**

```python
def encrypt(message, key, shift_direction):
    encrypted_message = ""
    for char in message:
        if char.isalpha():  # Check if the character is an alphabet
            if char.isupper():  # Check if the character is uppercase
                encrypted_char = chr((ord(char) - 65 + key * shift_direction) % 26 + 65)
                # Encrypt uppercase character by shifting its ASCII value by key * shift_direction
                # Wrap around using modulo 26 and add 65 to get the ASCII value of the encrypted character
                # ASCII values for uppercase letters range from 65 to 90
            else:
                encrypted_char = chr((ord(char) - 97 + key * shift_direction) % 26 + 97)
                # Encrypt lowercase character by shifting its ASCII value by key * shift_direction
                # Wrap around using modulo 26 and add 97 to get the ASCII value of the encrypted character
                # ASCII values for lowercase letters range from 97 to 122
        elif char.isdigit():  # Check if the character is a number
            encrypted_char = chr((ord(char) - 48 + key * shift_direction) % 10 + 48)
            # Encrypt the digit by shifting its ASCII value by key * shift_direction
            # Wrap around using modulo 10 and add 48 to get the ASCII value of the encrypted character
            # In ASCII, the decimal values 48 to 57 correspond to the characters '0' to '9'
        elif char.isspace():  # Check if the character is a space
            encrypted_char = char  # Leave the space character unchanged
        else:
            encrypted_char = char
        encrypted_message += encrypted_char  # Append the encrypted character to the encrypted message
    return encrypted_message

def decrypt(encrypted_message, key, shift_direction):
    decrypted_message = ""
    for char in encrypted_message:
        if char.isalpha():  # Check if the character is an alphabet
            if char.isupper():  # Check if the character is uppercase
                decrypted_char = chr((ord(char) - 65 - key * shift_direction) % 26 + 65)
                # Decrypt uppercase character by shifting its ASCII value by key * shift_direction
                # Wrap around using modulo 26 and add 65 to get the ASCII value of the decrypted character
                # ASCII values for uppercase letters range from 65 to 90
            else:
                decrypted_char = chr((ord(char) - 97 - key * shift_direction) % 26 + 97)
                # Decrypt lowercase character by shifting its ASCII value by key * shift_direction
                # Wrap around using modulo 26 and add 97 to get the ASCII value of the decrypted character
                # ASCII values for lowercase letters range from 97 to 1221

        elif char.isdigit():  # Check if the character is a number
            decrypted_char = chr((ord(char) - 48 - key * shift_direction) % 10 + 48)
            # Decrypt the digit by shifting its ASCII value by key * shift_direction
            # Wrap around using modulo 10 and add 48 to get the ASCII value of the decrypted character
            # In ASCII, the decimal values 48 to 57 correspond to the characters '0' to '9'
        elif char.isspace():  # Check if the character is a space
            decrypted_char = char  # Leave the space character unchanged
        else:
            decrypted_char = char
        decrypted_message += decrypted_char  # Append the decrypted character to the decrypted message
    return decrypted_message

message = input("Enter the message to encrypt/decrypt: ")
key = int(input("Enter the key (a positive integer): "))
shift_direction = int(input("Enter the shift direction (1 for right shift, -1 for left shift): "))

encrypted_message = encrypt(message, key, shift_direction)
print("Encrypted message:", encrypted_message)

decrypted_message = decrypt(encrypted_message, key, shift_direction)
print("Decrypted message:", decrypted_message)
```

```python
import pyAesCrypt
from os import stat, remove

print("1. Data XOR Encryption and Decryption")
print("2. Data AES Encryption and Decryption")
a = int(input("Enter your choice: "))  # Prompt the user to enter their choice

if a == 1:  # If the user chooses XOR Encryption and Decryption
    print("1. Press 1 for Encryption")
    print("2. Press 2 for Decryption")
    b = int(input("Enter your choice: "))  # Prompt the user to choose between encryption and decryption

    if b == 1:  # If the user chooses encryption
        path = input(r'Enter path of data: ')  # Prompt the user to enter the path of the data file
        key = int(input('Enter Key for encryption of data: '))  # Prompt the user to enter the encryption key
        print('The path of file: ', path)
        print('Key for encryption: ', key)
        fin = open(path, 'rb')  # Open the file in binary mode for reading
        image = fin.read()  # Read the binary value of the file
        fin.close()
        image = bytearray(image)  # Convert the binary value to a bytearray
        for index, values in enumerate(image):
            image[index] = values ^ key  # XOR each byte with the encryption key
        fin = open(path, 'wb')  # Open the file in binary mode for writing
        fin.write(image)  # Write the encrypted data back to the file
        fin.close()
        print('Encryption Done Successfully!')

    elif b == 2:  # If the user chooses decryption
        path = input(r'Enter path of data: ')  # Prompt the user to enter the path of the data file
        print('*********************************************************')
        print('Note: Encryption key and Decryption key must be the same!')
        print('*********************************************************')
        key = int(input('Enter Key for encryption of data: '))  # Prompt the user to enter the decryption key
        print('The path of file: ', path)
        print('Key for Decryption: ', key)
        fin = open(path, 'rb')  # Open the file in binary mode for reading
        image = fin.read()  # Read the binary value of the file
        fin.close()
        image = bytearray(image)  # Convert the binary value to a bytearray
        for index, values in enumerate(image):
            image[index] = values ^ key  # XOR each byte with the decryption key
        fin = open(path, 'wb')  # Open the file in binary mode for writing
        fin.write(image)  # Write the decrypted data back to the file
        fin.close()
        print('Decryption Done Successfully!')
    else:
        print("Please enter the right choice!")

elif a == 2:  # If the user chooses AES Encryption and Decryption
    print("1. Press 1 for Encryption")
    print("2. Press 2 for Decryption")
    h = int(input("Enter the Choice: "))  # Prompt the user to choose between encryption and decryption
    jh = input("Enter the file location with extension:")  # Prompt the user to enter the file location with extension
    bufferSize = 64 * 1024  # Set the buffer size for encryption/decryption
    password = input("Enter the password: ")  # Prompt the user to enter the password
    koi = f'{jh}.aes'  # Append ".aes" to the file name for the encrypted file

    if h == 1:  # If the user chooses encryption
        with open(jh, "rb") as fIn:  # Open the file in binary mode for reading
            with open(koi, "wb") as fOut:  # Open the encrypted file in binary mode for writing
                pyAesCrypt.encryptStream(fIn, fOut, password, bufferSize)  # Encrypt the file using AES encryption
        print("Encryption Done Succcesfully!")

    elif h == 2:  # If the user chooses decryption
        encFileSize = stat(jh).st_size  # Get the size of the encrypted file
        print(encFileSize)
        aso = input("Decrypted file name (Make sure to put the extension of the format you want your decrypted file in!):")  # Prompt the user to enter the name of the decrypted file
        with open(jh, "rb") as fIn:  # Open the encrypted file in binary mode for reading
            try:
                with open(aso, "wb") as fOut:  # Open the decrypted file in binary mode for writing
                    pyAesCrypt.decryptStream(fIn, fOut, password, bufferSize, encFileSize)  # Decrypt the file using AES decryption
            except ValueError:
                print("Value Error: Please give a smaller buffer size block")  # Handle the case where the buffer size is too large
        print("Decryption Done Succcesfully!")
else:
    print("Please enter the right choice!")
```
