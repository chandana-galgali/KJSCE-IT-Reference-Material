EXPERIMENT-1:LCM

```python
def lcm(X, a, c, m, N):
  rand_no = []
  p = None
  seen = {}
  for i in range(N):
    X = (a * X + c) % m
    rand_no.append(X / m)
    if X / m in seen:
      p = i - seen[X / m]
      break
    else:
      seen[X / m] = i
  if p is None:
    print("No repeat found. Period is full sequence length.")
    p = N
  return rand_no, p
X = 1
a = 5
c = 1
m = 8
N = 20
print("Seed = ", X)
print("Multiplier = ", a)
print("Increment = ", c)
print("Modulus = ", m)
print("Number of Random Numbers Generated = ", N)
rand_no, p = lcm(X, a, c, m, N)
print("\nGenerated Random Numbers: (Scaled to [0, 1]): ", rand_no)
print("\nPeriod of the sequence: ", p)
```

EXPERIMENT-3:KS

```python
import numpy as np
from scipy.stats import ks_2samp
def lcm(X, a, c, m, N):
  rand_no = []
  for i in range(N):
    X = (a * X + c) % m
    rand_no.append(X / m)
  return rand_no
def ls(rand_no):
  alpha = 0.05
  ud = np.random.uniform(0, 1, len(rand_no))
  ks_stat, p_value = ks_2samp(rand_no, ud)
  print("\nKolmogorov-Smirnov Statistic: ", ks_stat)
  print("\nP-Value: ", p_value)
  if p_value < alpha:
      print("\nReject the null hypothesis: The random numbers are not uniformly distributed.")
  else:
      print("\nFail to reject the null hypothesis: The random numbers are uniformly distributed.")
  return
X = 123456789
a = 1103515245
c = 12345
m = 128
N = 500
rand_no = lcm(X, a, c, m, N)
print("\nGenerated Random Numbers: (Scaled to [0, 1]): ", rand_no)
ls(rand_no)
```

EXPERIMENT-3:CHI-SQUARE

```python
import numpy as np
from scipy.stats import chisquare
def lcm(X, a, c, m, N):
  rand_no = []
  for i in range(N):
    X = (a * X + c) % m
    rand_no.append(X / m)
  return rand_no
def cs(rand_no):
  bins = 10
  alpha = 0.05
  obs_freq, bin_edges = np.histogram(rand_no, bins = bins, range=(0, 1))
  exp_freq = np.full_like(obs_freq, len(rand_no)/bins)
  chi2stat, p_value = chisquare(obs_freq, exp_freq)
  print("\nChi-Square Statistic: ", chi2stat)
  print("\nP-Value: ", p_value)
  if p_value < alpha:
      print("\nReject the null hypothesis: The random numbers are not uniformly distributed.")
  else:
      print("\nFail to reject the null hypothesis: The random numbers are uniformly distributed.")
  return
X = 123456789
a = 1103515245
c = 12345
m = 128
N = 500
rand_no = lcm(X, a, c, m, N)
print("\nGenerated Random Numbers: (Scaled to [0, 1]): ", rand_no)
cs(rand_no)
```
